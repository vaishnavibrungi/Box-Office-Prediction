### Feature Engineering

```python
#https://www.interviewqs.com/ddi-code-snippets/extract-month-year-pandas
#Create a new column 'release_year' to get year movie was released
train['release_year']=pd.DatetimeIndex(train['release_date']).year
test['release_year']=pd.DatetimeIndex(test['release_date']).year
```

```python
#https://stackoverflow.com/questions/25146121/extracting-just-month-and-year-separately-from-pandas-datetime-column
#Create a new column 'release_month' to get month movie was released
train['release_month']=pd.DatetimeIndex(train['release_date']).month
test['release_month']=pd.DatetimeIndex(test['release_date']).month
```

```python
#https://stackoverflow.com/questions/47970561/splitting-a-column-of-strings-and-counting-the-number-of-words-with-pandas
#Create a new column 'title_words_count' and get number of words in title each row
train['title_words_count'] = train['title'].str.split().str.len()
test['title_words_count'] = test['title'].str.split().str.len()
```

```python
#https://docs.python.org/3/library/re.html
#https://stackoverflow.com/questions/62006512/lambda-function-and-regular-expression
#create 'genres_types' and get names of genres
train['genres_types']=train['genres'].apply(lambda x:re.findall("'name': \'(.+?)\'", x))
test['genres_types']=test['genres'].apply(lambda x:re.findall("'name': \'(.+?)\'", x))
```

```python
#get 'genres_types' in a list format
train['genres_types'].tolist()
```

```python
#get 'genres_types' in a list format
test['genres_types'].tolist()
```

```python
#https://www.w3resource.com/python-exercises/list/python-data-type-list-exercise-183.php
#get unique values in list of lists
def unique_values_in_list(lists):
    result=set(x for l in lists for x in l) 
    return list(result)
```

```python
unique_genre_list_train = unique_values_in_list(train['genres_types'].tolist())
unique_genre_list_test = unique_values_in_list(test['genres_types'].tolist())
```

```python
#find length of list
len(unique_genre_list_test)
```

```python
feature_genere = []; #create an empty list
for genree in unique_genre_list_train: #collect 'genre' features and add empty columns
    train['genre_' + genree] = 0
    test['genre_' + genree] = 0
    feature_genere.append('genre_' + genree) 
```

```python
#https://stackoverflow.com/questions/23330654/update-a-dataframe-in-pandas-while-iterating-row-by-row
#for each row add 1 for corresponding 'genres_types'
for i, row in train.iterrows():
    for genere in row['genres_types']:
        train.at[i, 'genre_' + genere] = 1
        
for i, row in test.iterrows():
    for genere in row['genres_types']:
        test.at[i, 'genre_' + genere] = 1        
```

## Data Modeling

```python
#Set X and y
features = ['budget','popularity', 'release_month','release_year','title_words_count',
            'runtime'] + feature_genere #concatenate 
X=train[features]
y=train['revenue']
```

```python
X_testdata=test[features] 
```

```python
#Use train_test_split to generate your train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Get the shapes of train and test data
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Data Scaling

```python
# Scale the data
sc = StandardScaler()
Z_train = sc.fit_transform(X_train)
Z_test = sc.transform(X_test)
test_sc=sc.transform(X_testdata)
```

```python
# Get the shapes of train and test data
print(f'Z_train shape is: {Z_train.shape}')
print(f'y_train shape is: {y_train.shape}')
print(f'Z_test shape is: {Z_test.shape}')
print(f'y_test shape is: {y_test.shape}')
```

### Null Model

```python
#create y_null
y_null=np.zeros_like(y_train)
y_null.fill(y_train.mean()) 
y_null[0:5]
```

```python
#get null MSE
mean_squared_error(y_train,y_null)
```

```python
#get null RMSE
mean_squared_error(y_train,y_null,squared=False)
```

- By predicting the average every time, we are wrong or off by 140027435.56356192 on average.

### Linear Regression

```python
#Instaniate and fit LR
ols = LinearRegression()
ols.fit(Z_train, y_train)
```

```python
# Model score on the training and test data
print(" OLS ".center(18, "="))
print(ols.score(Z_train, y_train))
print(ols.score(Z_test, y_test))
print()
```

```python
#Generate predictions from LR model
ols.predict(Z_train) 
```

```python
#Generate predictions from LR model
ols.predict(Z_test)
```

```python
# Create predictions
y_preds=ols.predict(Z_train)
y_preds_test=ols.predict(Z_test)
```

```python
mean_squared_error(y_true=y_test, y_pred=y_preds_test, squared=False) # RMSE for the LR Model
```

### RidgeCV Regression

```python
# Set up a list of ridge alphas to check.
# np.logspace generates 10 values equally between 0 and 2,
# then converts them to alphas between 10^0 and 10^2.
r_alphas = np.logspace(0, 2, 10)
# Cross-validate over our list of ridge alphas.
ridge_cv = RidgeCV(alphas=r_alphas, scoring='r2', cv=5)
```

```python
# Fit model using best ridge alpha!
ridge_cv.fit(Z_train, y_train)
```

```python
# Here is the optimal value of alpha
ridge_cv.alpha_
```

```python
print(" Ridge ".center(18, "="))
print(ridge_cv.score(Z_train, y_train))
print(ridge_cv.score(Z_test, y_test))
```

```python
# Generate predictions from Ridge model
ridge_preds=ridge_cv.predict(Z_train)
ridge_preds_test=ridge_cv.predict(Z_test)
```

```python
#RMSE for the  Ridge Model
mean_squared_error(y_true=y_test, y_pred=ridge_preds_test, squared=False)
```

### LassoCV Regression

```python
# Cross-validate over Lasso alphas and fit model using best lasso alpha
lasso_cv = LassoCV(n_alphas=1000, cv=5)
lasso_cv.fit(Z_train, y_train);
```

```python
# Here is the optimal value of alpha
lasso_cv.alpha_ 
```

```python
print(" LASSO ".center(18, "="))
print(lasso_cv.score(Z_train, y_train))
print(lasso_cv.score(Z_test, y_test))
```

```python
# Generate predictions from Lasso model
lasso_preds=lasso_cv.predict(Z_train)
lasso_preds_test=lasso_cv.predict(Z_test)
```

```python
#RMSE for the  Lasso Model
mean_squared_error(y_true=y_test, y_pred=lasso_preds_test,squared=False) 
```

```python
#Convert 'RidgeCV' coefficients into dataframe
coef_df = pd.DataFrame([ridge_cv.coef_, features], index = ['Coefficients', 'Features']).T
coef_df = coef_df.set_index('Features')

coef_df = coef_df.sort_values(by='Coefficients',ascending=False)
coef_df.head(10) 
```

Everything else being held equal, budget increases the revenue of the movie  by about $86526819
